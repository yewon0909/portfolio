```python
from gensim.models import Word2Vec
import pandas as pd
import numpy as np

df = pd.read_csv('./data/re_data.csv')
corpus = []

for text in df['re_text']:
    rwords = []
    words = text.split(' ')
    for w in words: 
        if w != '': #!= í•´ë‹¹í•˜ì§€ ì•Šì„ ë•Œê°€ íŠ¸ë£¨ !notì´ëž‘ ë˜‘ê°™ì€ ì˜ë¯¸ìž„ ì›ëž˜ëŠ” ==
            rwords.append(w)
    corpus.append(rwords)
    
for text in df['re_title']:
    rwords = []
    words = str(text).split(' ') #floatì˜¤ë¥˜ strí•¨ìˆ˜ë¡œ ë¬¸ìžì—´ë¡œ ë§Œë“¤ì–´ì£¼ê¸°
    for w in words: 
        if w != '':
            rwords.append(w)
    corpus.append(rwords)
    
corpus
```

```python
model = Word2Vec(sentences=corpus, vector_size=100, window=5, min_count=1, workers=4, sg=1, hs=1, epochs=30)
model.save('./model/model1.model')
```
