```python
from gensim.models import Word2Vec
import pandas as pd
import numpy as np

df = pd.read_csv('./data/re_data.csv')
corpus = []

for text in df['re_text']:
    rwords = []
    words = text.split(' ')
    for w in words: 
        if w != '': #!= 해당하지 않을 때가 트루 !not이랑 똑같은 의미임 원래는 ==
            rwords.append(w)
    corpus.append(rwords)
    
for text in df['re_title']:
    rwords = []
    words = str(text).split(' ') #float오류 str함수로 문자열로 만들어주기
    for w in words: 
        if w != '':
            rwords.append(w)
    corpus.append(rwords)
    
corpus
```

```python
model = Word2Vec(sentences=corpus, vector_size=100, window=5, min_count=1, workers=4, sg=1, hs=1, epochs=30)
model.save('./model/model1.model')
```

```python
model = Word2Vec.load('./model/model1.model')
sims = model.wv.most_similar('memory', topn=10) #0-1 사이의 값으로 나옴
sims
```
