```python
import pandas as pd

df = pd.read_excel('./data/re_data.xlsx')
df
```

```python
import re
import nltk

stop_words = nltk.corpus.stopwords.words('english')
stop_words2 = ['the','an','first','second','in','using','test','may','one','also','two', 'end', 'least', 'method', 'methods', 'includes', 'including', 'include']

stop_words = set(stop_words + stop_words2)
re_titles = []

for title in df['title']:
    title = title.strip().lower()
    re_title = re.sub('[;""-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', title)
    words = re_title.split(' ')
    words = [w for w in words if w not in stop_words]
    re_title = ' '.join(words)
    re_titles.append(re_title.strip())
    
df['re_title'] = re_titles
df
```

```python
a = 'Priority 2019-01-12 • Filed 2019-03-08 • Publi'
a_list = a.split('•')
aa_list = a_list[1].strip().split(' ')
aa_list
```

```python
redates = []
for date in df['date']:
    dates = date.split('•')
    redate = dates[0].strip().split(' ')
    redates.append(redate[1])
    
df['re_date'] = redates
df
```

```python
re_texts = []

for text in df['f_text']:
    text = text.strip().lower()
    text = re.sub('[;""-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', text)
    words = text.split(' ')
    words = [w for w in words if w not in stop_words]
    re_text = ' '.join(words)
    re_texts.append(re_text.strip())
    
df['re_text'] = re_texts
df
```

```python
df.to_csv('./data/re_data!.csv', index = False)
df.to_excel('./data/re_data!.xlsx', index = False)
```
